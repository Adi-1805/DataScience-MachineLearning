#### Missing Values

````python
## Here we will check the percentage of NaN values present in each feature
## Step 1: make a list of features with missing values
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]
````

````python
## Step 2: print the feature name and the percentage of missing values
for feature in features_with_na:
    print(feature, np.round((dataset[feature].isnull().sum()/1460)*100, 4), '% of values are missing')
````

##### Since there are many missing values, we need to find the relationship between missing values and SalesPrice.
We will plot some graphs to analyze the same

````python
data = dataset.copy()
for feature in features_with_na:

    # Make a feature that occupies the value 1 if the observation was missing or zero otherwise
    data[feature] = np.where(data[feature].isnull(), 1, 0)

data.head()
     

````

In the above table/dataset (which is the copy of our original dataset), for each feature in the list, this line creates a new binary column in the data DataFrame where:
- 1 indicates that the value was missing.
- 0 indicates that the value was present.


````python
for feature in features_with_na:
    # Calculating the mean SalePrice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
````

Evidently, there is a relation between the missing values and the dependent variable. So we need to replace these NaN values with something meaningful which we will do in the Feature Engineering section.

#### Numerical Variables

````python
# list of numerical variables
# numerical features are categorized as those that do not have an objectType value (hence !='O')
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']

print('Number of numerical variables: ', len(numerical_features))

# visualise the numerical variables
dataset[numerical_features].head()
````

#### Temporal Variables(Eg: Datatime Variables)
From the Dataset, we have 4-year variables. We have extracted information from the dataset variables like number of years or number of days. One example of this specific scenario would be, the difference between the year the house was built and the year the house was sold. We will be performing this analysis in the Feature Engineering which is he next video.

````python
# list of variables that contain year information
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_feature
````

````python
# let us explore the content of these year variables
for feature in year_feature:
    print(feature, dataset[feature].unique())
````

````python
# let us analyze the 'Temporal Datetime Variables'
# We will check whether there is a relation between the year the house was built and the year the house was sold
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs YearSold')
````

Here, due to inflammation in house prices (which overrides other factors involving the market rate), the house price should increase as the year sold increases. But that is not the case, so we'll analyze other year_features.

````python
year_feature
````

````python
for feature in year_feature:
    if feature != 'YrSold':
        data = dataset.copy()
        # we will capture the difference between year_variable and the year in which the house was sold
        data[feature] = data['YrSold'] - data[feature]

        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        
        plt.ylabel('SalePrice')
        plt.show()
````

#### From each of the above, we can conclude the following:
1. As the difference between the year the house was built and sold increases, the selling price decreases.
    - New Houses tend to sell at higher prices.
2. In the YearRemodeling graph, the price evens out at the bottom.
    - Remodeling increases the selling price.
3. Building garage with the new house gets you a higher selling price.


````python
## Numerical features are of two types:
# 1. Continuous variable
# 2. Discreet variable

# In our dataset, we will consider those features as discreet where the number of unique values is less than 25 
# and that are not part of the 'year_feature' columns and 'id' column
discreet_feature = [feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature +['Id']]
print("Number of discreet features are: {}".format(len(discreet_feature)))
discreet_feature
````

````python
dataset[discreet_feature].head()
````

````python
## Finding relationship between them and SalePrice
for feature in discreet_feature:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
````

````python
# Setting a color palette
palettes = sns.color_palette("rainbow", len(discreet_feature))

# Finding relationship between discreet features and SalePrice
for i, feature in enumerate(discreet_feature):
    data = dataset.copy()
    
    # Create the plot
    median_values = data.groupby(feature)['SalePrice'].median()
    ax = median_values.plot.bar(color=palettes[i])
    
    # Set labels and title
    ax.set_xlabel(feature)
    ax.set_ylabel('SalePrice')
    ax.set_title(feature)
    
    # Show the plot
    plt.show()
````

````python
## We will be using logarithmic transformation
palette = sns.color_palette("rainbow", len(continuous_feature))
for i,feature in enumerate(continuous_feature):
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass 
    else:
        data[feature] = np.log(data[feature])
        data['Salerice'] = np.log(data['SalePrice'])
        sns.histplot(data[feature],bins = 30, kde=True, color = palette[i])
        plt.xlabel(feature)
        plt.ylabel('Sales Price')
        plt.title(feature)
        plt.show()
````

### NOTE: Look at the graph outputs of the last two cells, and notice that the skewed distribution is converted to normal distribution.

````python
## We will plot scatter plots to see the correlation between these continuous_feature and SalePrice 

for feature in continuous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass 
    else:
        data[feature] = np.log(data[feature])
        data['Salerice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sales Price')
        plt.title(feature)
        plt.show()
````


### Code Explanaon
 
The provided code snippet applies logarithmic transformations to continuous features in a dataset and then creates scatter plots to visualize the relationship between the transformed features and the target variable, `SalePrice`. Here's a step-by-step explanation of how the code works:on

```python
for feature in continuous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass 
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sales Price')
        plt.title(feature)
        plt.show()
```

1. **Iteration Over Continuous Features**:
   - `for feature in continuous_feature:`: This loop iterates over each feature in the list of continuous features.

2. **Data Copy**:
   - `data = dataset.copy()`: A copy of the original dataset is created to ensure that the original data remains unchanged during the transformation and plotting.

3. **Check for Zero Values**:
   - `if 0 in data[feature].unique(): pass`: The code checks if the feature contains any zero values. The logarithm of zero is undefined, so if any zeros are present, the code uses `pass` to skip the transformation and plotting for that feature.

4. **Logarithmic Transformation**:
   - `data[feature] = np.log(data[feature])`: This line applies the natural logarithm transformation to the current feature.
   - `data['SalePrice'] = np.log(data['SalePrice'])`: This line applies the natural logarithm transformation to the target variable, `SalePrice`.

5. **Scatter Plot Creation**:
   - `plt.scatter(data[feature], data['SalePrice'])`: Creates a scatter plot of the transformed feature against the transformed `SalePrice`.
   - `plt.xlabel(feature)`: Sets the label for the x-axis to the name of the current feature.
   - `plt.ylabel('Sales Price')`: Sets the label for the y-axis to 'Sales Price'.
   - `plt.title(feature)`: Sets the title of the plot to the name of the current feature.
   - `plt.show()`: Displays the scatter plot.

### Purpose of the Code

1. **Handling Skewed Data**:
   - Logarithmic transformation is often used to handle skewed data. It compresses the range of the data and can help in stabilizing the variance and making the data more normally distributed.

2. **Visualizing Relationships**:
   - The scatter plots created after the transformation help in visualizing the relationship between the continuous features and the target variable, `SalePrice`. By transforming both the feature and the target variable, the code aims to linearize the relationship, making patterns more apparent and easier to interpret.

3. **Data Preprocessing**:
   - This transformation is a part of data preprocessing, which can improve the performance and interpretability of regression models. Many models, such as linear regression, assume normally distributed residuals, and applying logarithmicontaining zero values are skipped to avoid undefined logarithmic transformations.
