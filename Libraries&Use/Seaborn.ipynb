```python
import seaborn as sns
df = sns.load_dataset("tips") #"tips" is a sample data set present in sns
df.head()
```

#### Correlation with Heatmap
A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete or event types. It is essential in Feature Selection.
- Correlation can only be measured in data with values, not categorical data.

```python
df.dtypes
```

```python
# making a sub-dataframe containing all the non-category type values to calculate correlation
subdf = df[['total_bill', 'tip', 'size']]
subdf.head()
```

```python
subdf.corr()
```

```python
sns.heatmap(subdf.corr())
```

### JointPlot
A joint plot allows to study the relation between 2 numeric variables. The central chart displays their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot.

```python
sns.jointplot(x='tip', y ='total_bill', data=subdf, kind='hex')
```

```python
sns.jointplot(x='tip', y ='total_bill', data=subdf, kind='reg')
# reg -> regression. The plot shows the regression line.
```

#### Pair plot
A pair plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pair plots are just elaborations on this, showing all variables with all the other variables.

```python
sns.pairplot(subdf)
# shows the plot gird with plot between all the non-categorical variables in the dataset
# a pair plot of one variable with itself is a histogram because it shows the variable ditibution not the correlation as the correlation is obviously one.
# pair plots of all the variables with a different variable is a scatter plot showing correlation
```

```python
df['smoker'].value_counts()
```

```python
sns.pairplot(df,hue='smoker')
```

#### Box Plot
A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot(data = df, x='smoker',y='total_bill')
```

```python
sns.boxplot(data = df, x='day',y='tip', palette ='rainbow')
```

```python
sns.boxplot(data = subdf, oreint ='v')
```
