## MatplotLib Tutorial
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Marplotlib are:
- Generally easy to get started for simple plots.
- Support for custom labels and texts
- Great control of every element in a figure
- High-quality output in many formats
- Very customizable in general

```python
import matplotlib.pyplot as plt
%matplotlib inline 
#if you don't include this inline statement, you will have to use the alias name with '.show()' to show the the actual graph
import numpy as np
```

```python
## Simple Examples
x = np.arange(0,10)
y = np.arange(11,21)
a = np.arange(40,50)
b = np.arange(50,60)
```

#### Plotting using matplotlib

```python
## plt scatter
plt.scatter(x,y,color='yellow')
# A scatter plot of *y* vs. *x* with varying marker size and/or color.
#Parameters
# x, y : float or array-like, shape (n, ). The data positions.
# s : float or array-like, shape (n, ), optional. The marker size in points**2 (typographic points are 1/72 in.).
# Default is ``rcParams['lines.markersize'] ** 2``.
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
y = x*x
```

```python
## plt plot
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
# .plot() -> uses indexable values eg., list, tuple, array, dict as first and second parameter and the later arguments are for graph appearance
# >>> plot(x, y, 'go--', linewidth=2, markersize=12)
# >>> plot(x, y, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12)

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Diagram')
```

```python
### creating subplots
p = np.linspace(0,100,100)
plt.subplot(3,1,1)
q = p*p
plt.plot(p,q,'r.',linewidth=1,markersize=2)
plt.subplot(3,1,2)
q = p**3
plt.plot(p,q,'g*',linewidth=1,markersize=2)
plt.subplot(3,1,3)
q = np. sin(p)
plt.plot(p,q,linewidth=1,markersize=2)
```

```python
# PIE GRAPH
labels = 'Python',' C++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold', 'yellowgreen','lightcoral','lightskybule']
explode = (0.1,0,0,0) #explode 1st slice
#plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow= True)
plt.axis('equal')
plt.show()
```
